```python
from pyspark.sql import SparkSession
from pyspark.sql import Row

data_employee = [
    Row(employee_id=1, name='Alice', city='New York'),
    Row(employee_id=2, name='Bob', city='Los_Angeles'),
    Row(employee_id=3, name='Charlie', city='Chicago'),
    Row(employee_id=4, name='David', city='Houston'),
    Row(employee_id=5, name='Eva', city='Phoenix')
]

# 2. Define Sample Salary info
data_salary = [
    Row(employee_id=1, salary=70000, deprt='HR'),
    Row(employee_id=2, salary=80000, deprt='Engineering'),
    Row(employee_id=3, salary=65000, deprt='Finance'),
    Row(employee_id=6, salary=90000, deprt='Marketing'),
    Row(employee_id=7, salary=75000, deprt='Sales')
]

data_employee=spark.createDataFrame(data_employee)
data_salary=spark.createDataFrame(data_salary)
data_employee.createOrReplaceTempView("data_employee")
data_salary.createOrReplaceTempView("data_salary")
display(data_employee)
display(data_salary)
```

```python
df_inner=data_employee.join(data_salary,on='employee_id',how='inner')
display(df_inner)
```

```python
%sql
select e.employee_id,e.name,e.city,s.salary,s.deprt
from data_employee e inner join data_salary s on e.employee_id=s.employee_id;
```

```python
df_inner=data_employee.join(data_salary,on='employee_id',how='outer')
display(df_inner)
```

```python
%sql
select e.employee_id,e.name,e.city,s.salary,s.deprt from 
data_employee e FULL OUTER JOIN data_salary s on e.employee_id=s.employee_id
```

```python
df_inner=data_employee.join(data_salary,on='employee_id',how='left_outer')
display(df_inner)
```

```python
df_inner=data_employee.join(data_salary,on='employee_id',how='right_outer')
display(df_inner)
```
